```python
def boxplot_iris(feature_names,dataset):
  i=1
  plt.figure(figsize=(11,9))
  for col in feature_names:
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis='both',left=True,
                    top=False,right=False,
                    bottom=True,labelleft=False,
                    labeltop=False,labelright=False,
                    labelbottom=False)
    dataset[col].plot(kind='box',subplots=True,sharex=False)
    plt.title(col)
    i+= 1
  plt.show()
```

```python
boxplot_iris(iris.columns[:-1],iris)
```

```python
def histogram_iris(feature_names,dataset):
  i=1
  plt.figure(figsize=(11,9))
  for col in feature_names:
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis='both',left=True,
                    top=False,right=False,
                    bottom=True,labelleft=False,
                    labeltop=False,labelright=False,
                    labelbottom=False)
    dataset[col].hist()
    plt.title(col)
    i+= 1
  plt.show()
```

```python
histogram_iris(iris.columns[:-1],iris)
```

```python
def piecher_iris(features_names, target, dataset):
  i=1
  plt.figure(figsize=(11,9))
  for colName in [target]:
    labels=[];size=[];
    df = dataset.groupby(colName).size()
    for key in df.keys():
      labels.append(key)
      size.append(df[key])
    plt.subplot(2,2,i)
    plt.axis('on')
    plt.tick_params(axis='both',left=True,
                    top=False, right = False,
                    bottom=True, labelleft=True,
                    labeltop=True, labelright=False,
                    labelbottom=False)
    plt.pie(size, labels=labels, autopct='%1.1f%%',
            shadow=True,startangle=140)
    plt.axis('equal')
    i+=1
  plt.show()
```

```python
piecher_iris(iris.columns[:-1],iris.species,iris)
```
